### 1. Instalando virtualbox

Descargamos la versión para windows de la web de [virtualbox](https://www.virtualbox.org/wiki/Downloads) haciendo click en *Windows hosts*:

<img src="img/virtualbox_1.png" style="float:left">
<div style="clear:both"></div><br>

Después descargamos el extensión pack pra virtualbox haciendo click en *All supported platforms*:

<img src="img/virtualbox_2.png" style="float:left">
<div style="clear:both"></div><br>

Tras descargarlo, lo ejecutamos y pulsamos el botón *Next*:

<img src="img/virtualbox_3.png" style="float:left">
<div style="clear:both"></div><br>

Nuevamente pulsamos el botón *Next*:

<img src="img/virtualbox_4.png" style="float:left">
<div style="clear:both"></div><br>

Seguimos pulsando el botón *Next*:

<img src="img/virtualbox_5.png" style="float:left">
<div style="clear:both"></div><br>

A continuación, pulsamos el botón *Yes*:

<img src="img/virtualbox_6.png" style="float:left">
<div style="clear:both"></div><br>

Finalmente, pulsamos el botón install:

<img src="img/virtualbox_7.png" style="float:left">
<div style="clear:both"></div><br>

para que comience la instalación:

<img src="img/virtualbox_8.png" style="float:left">
<div style="clear:both"></div><br>

al finalizar la instalación, pulsamos el botón*Finish*:

<img src="img/virtualbox_9.png" style="float:left">
<div style="clear:both"></div><br>

para ejecutar virtualbox:

<img src="img/virtualbox_10.png" style="float:left">
<div style="clear:both"></div><br>

Una vez instalado virtualbox, instalamos el *extension pack*. Para ello, lo ejecutamos haciendo doble click:

<img src="img/virtualbox_11.png" style="float:left">
<div style="clear:both"></div><br>

Pulsamos el botón *Install* para continuar:

<img src="img/virtualbox_12.png" style="float:left">
<div style="clear:both"></div><br>

Entendiendo en el tutorial que se acepta la licencia, pulsamos el botón *I Agree*:

<img src="img/virtualbox_13.png" style="float:left">
<div style="clear:both"></div><br>

Una vez que se instale, pulsamos el botón *OK*:

<img src="img/virtualbox_14.png" style="float:left">
<div style="clear:both"></div><br>

Ya tenemos virtualbox y pack de extensiones instalado. Con el fin de seguir mejor los ejemplos de este manual, vamos a configurar una red en [NAT](https://es.wikipedia.org/wiki/Traducci%C3%B3n_de_direcciones_de_red). Para ello, seleccionamos la opción *Preferences...* del menú *File*:

<img src="img/virtualbox_15.png" style="float:left">
<div style="clear:both"></div><br>

En el menú de opciones, seleccionamos *Network* en el menú de la izquierda y pulsamos el botón que tiene un signo **+**:

<img src="img/virtualbox_16.png" style="float:left">
<div style="clear:both"></div><br>

Se creará un nuevo *NatNetwork*. Ahora pulsamos en el botón con el icono engranaje para configurarlo:

<img src="img/virtualbox_17.png" style="float:left">
<div style="clear:both"></div><br>

Establecemos la red que queremos -para seguir los ejemplos del libro se ha elegido la red *192.168.119.0/24*- y pulsamos el botón *OK*:

<img src="img/virtualbox_18.png" style="float:left">
<div style="clear:both"></div><br>

### Metasploitable 2

#### Instalando metasploitable 2

Descargamos [Metasploitable 2](https://sourceforge.net/projects/metasploitable/files/Metasploitable2/)

<img src="img/metasploitable2_1.png" style="float:left">
<div style="clear:both"></div><br>

Descomprimimos la carpeta para ver los archivos. De todos ellos, necesitaremos *Metasploitable.vmdk*:

<img src="img/metasploitable2_2.png" style="float:left">
<div style="clear:both"></div><br>

Vamos a virtualbox y pulsamos el botón *New*:

<img src="img/metasploitable2_3.png" style="float:left">
<div style="clear:both"></div><br>

Le damos un nombre a la máquina, por ejemplo, Metasploitable. Después seleccionamos *Linux* en el menú desplegable *Type* y seleccionamos *Ubuntu (32-bit)* en el menú desplegable *Version*. Finalmente pulsamos el botón *Next*:

<img src="img/metasploitable2_4.png" style="float:left">
<div style="clear:both"></div><br>

Dejamos la memoria que viene por defecto y pulsamos el botón *Next*:

<img src="img/metasploitable2_5.png" style="float:left">
<div style="clear:both"></div><br>

A continuación, seleccionamos la opción *Use an existing virtual hard disk file* y pulsamos el icono de la carpeta para buscarlo:

<img src="img/metasploitable2_6.png" style="float:left">
<div style="clear:both"></div><br>

Seleccionamos el archivo *Metasploitable.vmdk* que está en la carpeta que hemos descomprimido y pulsamos el botón *Open*:

<img src="img/metasploitable2_7.png" style="float:left">
<div style="clear:both"></div><br>

Pulsamos el botón *Create* para finalizar:

<img src="img/metasploitable2_8.png" style="float:left">
<div style="clear:both"></div><br>

Veremos que se ha creado la máquina virtual *Metasploitable* en virtualbox:

<img src="img/metasploitable2_9.png" style="float:left">
<div style="clear:both"></div><br>

Antes de iniciarla, con la máquina seleccionada en el panel de la izquierda, pulsamos el botón *Settings*:

<img src="img/metasploitable2_10.png" style="float:left">
<div style="clear:both"></div><br>

En la pestaña *Network*, seleccionamos la opción *NAT Network* del menú *Attached to* para unir la máquina a nuestra red NAT:

<img src="img/metasploitable2_11.png" style="float:left">
<div style="clear:both"></div><br>

finalmente pulsamos el botón *OK*:

<img src="img/metasploitable2_12.png" style="float:left">
<div style="clear:both"></div><br>

#### Configurando el servicio snmp

Iniciamos la máquina *Metasploitable* pulsando el botón *Start*:

<img src="img/metasploitable2_13.png" style="float:left">
<div style="clear:both"></div><br>

Cuando apareca la pantalla de logín, vemos que nos indica que las credenciales son:

  * **Usuario:** msfadmin
  * **Contraseña:** msfadmin
  
<img src="img/metasploitable2_14.png" style="float:left">
<div style="clear:both"></div><br>

Nos logamos en la máquina usando esas credenciales:

<img src="img/metasploitable2_15.png" style="float:left">
<div style="clear:both"></div><br>

Ahora editaremos la configuración del servicio snmp para que, en vez de sólo estar a la escucha en localhost, lo este en la red NAT que hemos creado. Para editar la configuración del servicio snmp, editamos el fichero */etc/default/snmpd* con cualquier editor de texto. En este caso usamos [nano](https://www.nano-editor.org/):

<img src="img/metasploitable2_16.png" style="float:left">
<div style="clear:both"></div><br>

Autorizamos la operación con la contraseña *msfadmin*:

<img src="img/metasploitable2_17.png" style="float:left">
<div style="clear:both"></div><br>

Una vez se abrá el fichero, tenemos que localizar la dirección IP del localhost (127.0.0.1) de la línea que empieza por *SNMPDOPTS*:

<img src="img/metasploitable2_18.png" style="float:left">
<div style="clear:both"></div><br>

Una vez localizada, borramos la dirección IP. La línea debería quedar así:

<img src="img/metasploitable2_19.png" style="float:left">
<div style="clear:both"></div><br>

Una vez modificada la línea pulsamos la combinación de teclas *Ctrl+X* para salir:

<img src="img/metasploitable2_20.png" style="float:left">
<div style="clear:both"></div><br>

Pulsamos la tecla **Y** para confirmar que queremos guardar los cambios que hemos realizado al fichero:

<img src="img/metasploitable2_21.png" style="float:left">
<div style="clear:both"></div><br>

Finalmente pulsamos la tecla *Intro* para confirmar y salir:

<img src="img/metasploitable2_22.png" style="float:left">
<div style="clear:both"></div><br>

Una vez realizado los cambios en el fichero podemos iniciar el servicio snmp tecleando: *sudo /etc/init.d/snmpd start* y autorizando la operación con nuestra contraseña (*msfadmin*):

<img src="img/metasploitable2_23.png" style="float:left">
<div style="clear:both"></div><br>

El servicio quedará iniciado:

<img src="img/metasploitable2_24.png" style="float:left">
<div style="clear:both"></div><br>

Necesitaremos iniciar este servicio en la fase de enumeración. Para detener el servicio snmp tecleamos: *sudo /etc/init.d/snmpd stop*:

<img src="img/metasploitable2_25.png" style="float:left">
<div style="clear:both"></div><br>

Cuando terminemos de usar la máquina elegimos la opción *Close...* del menú *File*:

<img src="img/metasploitable2_26.png" style="float:left">
<div style="clear:both"></div><br>

Seleccionamos la opción *Power off the machine* y pulsamos el botón *OK*:

<img src="img/metasploitable2_27.png" style="float:left">
<div style="clear:both"></div><br>
